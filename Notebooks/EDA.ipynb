### Outliers en la variable `car_age`

Durante el análisis exploratorio se identificaron valores atípicos en la variable `car_age`, con una edad máxima de 110 años.  
Aunque este valor se encuentra fuera del rango típico de uso para la mayoría de los vehículos, no se considera un error de datos.

Este tipo de observaciones puede corresponder a autos clásicos o de colección que aún se encuentran en venta en el mercado.  
Por esta razón, los valores atípicos se conservan en el conjunto de datos para evitar la eliminación de información potencialmente relevante y mantener la diversidad del mercado representado.

“La conservación de estos valores permite un análisis más representativo del mercado real de vehículos usados.”

### Valores faltantes en la variable `odometer`

La variable `odometer` presenta aproximadamente un 15% de valores faltantes.
Dado que la distribución del kilometraje es asimétrica y contiene valores
extremos elevados, se decidió imputar los valores faltantes utilizando la
mediana, ya que es una medida robusta frente a outliers.

Esta decisión permite conservar la mayor cantidad de observaciones sin
distorsionar la distribución original de la variable.
supongo que:

0 km → posible error o auto nuevo

990,000 km → vehículos comerciales / trabajo

### Valores faltantes en la variable `paint_color`

La variable categórica `paint_color` presenta aproximadamente un 18% de valores
faltantes. Dado que el color no puede inferirse de forma confiable y que eliminar
las observaciones implicaría una pérdida significativa de información, se optó
por crear una categoría explícita denominada `unknown`.

Esta estrategia permite conservar todas las observaciones y evita introducir
sesgo al imputar un valor arbitrario.

### Tratamiento de valores faltantes en `is_4wd`

La variable `is_4wd` indica si el vehículo cuenta con tracción en las cuatro
ruedas (1 = sí). Los valores faltantes se interpretan como ausencia de esta
característica, práctica común en listados de vehículos donde solo se marca
cuando el atributo está presente.

Por esta razón, los valores nulos se imputaron con 0, indicando que el vehículo
no es 4WD.

### Creación de la variable `car_age`

Se creó la variable `car_age` como la diferencia entre el año de publicación
del anuncio (`date_posted`) y el año del modelo del vehículo (`model_year`).
Esta variable permite analizar el efecto de la antigüedad del vehículo sobre
el precio de venta.
